# Find similar words

Find spelling variants and OCR-errors

```python
import nbtext as nb
from nbtext import sorted_wildcardsearch
```



Use * as wildcard, and *factor* determines size of matching words, which gives a kind of Levenshtein-effekt. 


#### Look for OCR-errors

```python
sorted_wildcardsearch(
    {
        'word': 'kj*lig*d',   # search pattern
        'factor':3,           # factor determines length of result as difference from pattern
        'freq_lim':50,        # lowest frequency
        'limit':50            # number of hits
        
    })
```

#### Find inflections

```python
sorted_wildcardsearch(
    {
        'word': 'barn*',    # search pattern
        'factor':2,           # factor determines length of result as difference from pattern
        'freq_lim':50,        # lowest frequency
        'limit':50            # number of hits
        
    })
```

#### Find compounds

```python
sorted_wildcardsearch(
    {
        'word': 'sitron*',    # search pattern
        'factor':6,           # factor determines length of result as difference from pattern
        'freq_lim':50,        # lowest frequency
        'limit':50            # number of hits
        
    })
```
